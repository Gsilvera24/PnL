```python
# initial imports
import pandas as pd
from pathlib import Path
```

```python
#set file path
file_path = Path("./PnL.csv")
# create dataframe
df = pd.read_csv(file_path)
```

```python
#create a smaller dataframe where I will only look at these columns
table_one = df[['Trade Date','Ticker','Quantity','Currency','Price']].copy()
```

```python
# I want to create a new column named "Trade Value" since each trade has its own price which is different 
# than the prices listed on the other tables in the excel sheet.
table_one['Trade Value'] = table_one['Quantity'] * table_one['Price']
```

```python
#Check - this data frame has the trade values.
table_one
```

```python
# Since I no longer need the Price, i will drop t.
table_one.drop(['Price'], axis=1, inplace=True)
```
